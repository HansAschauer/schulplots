## Run schulplots in Jupyter Notebook
1. Install schulplots and write simple plot file to disk

```python
!pip install schulplots
!mkdir -p tmp_schulplots
plot = """figure:
  height: 8cm
  width: 11cm
axes_descriptors:
- axes:
    height: 6cm
    width: 4cm
    x_min: -1
    y_min: -3
    show_legend: false
  bottom: 1cm
  left: 1cm
  graphs:
  - function: exp(-x)
    plot_args: 
      ls: ":"
      color: orange
  - function: -exp(-x)
    plot_args: 
      ls: ":"
      color: orange
  - function: exp(-x) * cos(10*x)
- axes:
    height: 6cm
    width: 4cm
    x_min: -1
    y_min: -3
    show_legend: false
  bottom: 1cm
  left: 6cm
  graphs:
  - function: 2*sin(x)
    plot_args: 
      ls: ":"
      color: orange
  - function: -2*sin(x)
    plot_args: 
      ls: ":"
      color: orange
  - function: 2*sin(x) * cos(10*x)
"""
open("tmp_schulplots/plt.yaml","w").write(plot)
```

2. Do some magic required to get the coordinate systems right in Jupyter Notebooks

```python
%matplotlib inline
%config InlineBackend.print_figure_kwargs = {'bbox_inches':None}
```

3.  Generate the plot. It is also saved as `tmp_schulplots/plt.png`

```python
from schulplots.main import generate_plot
generate_plot("tmp_schulplots/plt.yaml", "tmp_schulplots/plt.png")
```

4. You can also run the `schulplots` binary

```python
! schulplots tmp_schulplots/plt.yaml
```

**Attention** This will remove your plot, so be cautious!

```python
! rm -rf tmp_schulplots
```
