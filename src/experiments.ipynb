```python
%reload_ext autoreload
%autoreload 2
from schulplots.saxes import SAxes
from schulplots.sgraph import SGraph, FillBetween, Discontinuity, DiscontinuityBelongsTo
from schulplots.figure_description import AxesWithGraphs, FigureDescription
from schulplots.types import Size
from schulplots.sfigure import SFigure
from schulplots.converter import converter
%matplotlib qt
```

```python

ax1 = SAxes(-1, -1, Size("8cm"), Size("10cm"), show_legend=True, unit=Size("2cm"))
g1 = SGraph("sin(x)", label= "$\sin(x)$")
g2 = SGraph("x**2", label="$x^2$")
g3 = SGraph("heaviside(x-2,0) * sin(x) + heaviside(2-x,1) * cos(x)", label="$g_3$",
            discontinuities=[
                Discontinuity(x0=2, belongs_to=DiscontinuityBelongsTo.SMALLER),
                Discontinuity(x0=1, belongs_to=DiscontinuityBelongsTo.NONE)
                ])
f1 = FillBetween(["sin(x)", "x**2"], condition="y1 >= y2",
                    plot_args=dict(alpha=0.2),
                    label="$\int_0^{x_0} sin(x) - x^2 \, dx$")
awg1 = AxesWithGraphs(ax1, [g2,g3], [f1], left=Size("1cm"), bottom=Size("1cm"))



ax2 = SAxes(-3, -2, Size("8cm"), Size("10cm"), show_legend=True)
g3 = SGraph("x**3", label="$x^3$")
g4 = SGraph("1/x", label="$1/x$")

f2 = FillBetween(["minimum(x**3, 1/x)", "0*x"], condition="x>0", label="A")
awg2 = AxesWithGraphs(ax2, [g3, g4],[f2], left=Size("9.5 cm"), bottom=Size("1cm"))

fdsc = FigureDescription(SFigure(width=Size("18 cm"), height=Size("12 cm")), [awg1, awg2])
fdsc.create_figure()
print(f1.get_closest_intersect(1))
print(converter.dumps(fdsc))
```
